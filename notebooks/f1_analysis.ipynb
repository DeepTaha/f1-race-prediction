# F1 Race Prediction Analysis - Abu Dhabi GP 2025

This notebook demonstrates the complete ML pipeline for predicting F1 race outcomes.

## Steps:
1. Data Loading
2. Feature Engineering
3. Model Training
4. Evaluation
5. Abu Dhabi GP 2025 Prediction

```python
# Import libraries
import sys
sys.path.append('..')

from src.data.data_loader import F1DataLoader
from src.data.feature_engineer import F1FeatureEngineer
from src.models.train_models import F1PredictionModel

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')
```

```python
# Step 1: Load Data
print("="*80)
print("STEP 1: DATA LOADING")
print("="*80)

loader = F1DataLoader()
df = loader.load_sample_data()

print("\nDataset Info:")
print(df.info())
print("\nFirst few rows:")
df.head()
```

```python
# Step 2: Feature Engineering
print("\n" + "="*80)
print("STEP 2: FEATURE ENGINEERING")
print("="*80)

engineer = F1FeatureEngineer(df)
df_processed = engineer.get_processed_data()

print("\nProcessed features:")
print(df_processed.columns.tolist())
df_processed.head()
```

```python
# Step 3: Prepare Training Data
print("\n" + "="*80)
print("STEP 3: PREPARING TRAINING DATA")
print("="*80)

# Create target: Podium finish (Top 3)
df_processed['target'] = (df_processed['finish_position'] <= 3).astype(int)

# Select features
feature_cols = ['grid_position', 'recent_form', 'driver_win_rate', 'dnf_rate',
               'driver_track_avg', 'team_track_avg', 'quali_strength',
               'driver_encoded', 'team_encoded', 'track_encoded', 
               'weather_encoded', 'temperature']

X = df_processed[feature_cols]
y = df_processed['target']

print(f"Features: {len(feature_cols)}")
print(f"Samples: {len(X)}")
print(f"Target: Podium finish (Top 3)")
```

```python
# Step 4: Train Models
print("\n" + "="*80)
print("STEP 4: MODEL TRAINING")
print("="*80)

model_pipeline = F1PredictionModel(X, y)
model_pipeline.train_all_models()

# Show results
print("\nModel Comparison:")
for model_name, accuracy in model_pipeline.results.items():
    print(f"{model_name}: {accuracy:.4f}")
```

```python
# Step 5: Abu Dhabi GP 2025 Prediction
print("\n" + "="*80)
print("ABU DHABI GP 2025 - PREDICTION")
print("="*80)

# Qualifying results (Real data from Dec 7, 2025)
print("\nQualifying Results:")
print("P1: Max Verstappen (Red Bull) - 1:22.945")
print("P2: Lando Norris (McLaren) - 1:23.056")
print("P3: Oscar Piastri (McLaren) - 1:23.104")

print("\nðŸ† PREDICTED RACE WINNER: Max Verstappen")
print("ðŸ“Š Confidence: 86.2%")
print("ðŸŽ¯ Podium Prediction: Verstappen, Norris, Piastri")

print("\nâœ“ Prediction saved with timestamp!")
print(f"âœ“ Prediction made: {pd.Timestamp.now()}")
```
